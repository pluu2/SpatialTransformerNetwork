Very slow implementation due to the dotproduct here. 

```python


#make an array of x,y coordinates 
def rotate(image1):  #This is an array of coordinates. 
  mask=np.zeros((784,3))
  i = 0
  for y in range(28):
    for x in range(28): 
      mask[i][0]=y-14
      mask[i][1]=x-14
      mask[i][2]=1
      i+=1


  for i in range(0,45):
    degrees=-i/10
    ident=np.array([[np.cos(degrees),np.sin(degrees),14],
                    [-np.sin(degrees),np.cos(degrees),14]
                                
                    ])
    #temp*=-1
    transform = np.dot(mask,ident.T)
    #print(transform)
    dimension=28

    image1=image1.flatten()
    new_image=np.zeros((dimension,dimension))
    for i in range(784): #unparallizable in this form. Very 
      if (np.int(transform[i][0])<= (dimension-1) and (np.int(transform[i][1])<=dimension-1)) and (np.int(transform[i][0])>= 0.0 and (np.int(transform[i][1])>=0.0)):
        #print(np.int(transform[i][0])) 
        #print(np.int(transform[i][1]))
        new_image[np.int(transform[i][0])][np.int(transform[i][1])] = image1[i]

    new_image=np.array(new_image)
    py.imshow(new_image.reshape(dimension,dimension))
    py.show()
```

```python
rotate(image1)
```
